```python
import pandas as pd
from pandas import DataFrame

import matplotlib.pyplot as plt
```

```python
# coffee sales csv saved as sales_df
sales_df = pd.read_csv('Coffe_sales.csv')
sales_df
```

```python
# df info
sales_df.info()
```

```python
#df shape
print(sales_df.shape)
```

```python
# check for duplicates in df 

duplicated_rows= sales_df[sales_df.duplicated()]
```

```python
print(duplicated_rows)
```

```python
sales_df
```

```python
# combine date and time column to one datetime column in dt format

sales_df['datetime'] = pd.to_datetime(
    sales_df['Date'] + ' ' + sales_df['Time'],
    format='mixed'
) 
```

```python
sales_df
```

```python
# check for empty columns

sales_df.info()
```

```python
# create hour, weekday, and month column using datetime column

sales_df['hour_check'] = sales_df['datetime'].dt.hour
sales_df['weekday_check'] = sales_df['datetime'].dt.day_name()
sales_df['month_check'] = sales_df['datetime'].dt.month_name()
```

```python
# use hour check column to check hour of day column make sure no unmmatch error

(sales_df['hour_check'] == sales_df['hour_of_day']).all()

```

```python
# use weekday check column to check weekday column make sure no unmmatch error


(sales_df['weekday_check'] == sales_df['Weekday']).all()
```

```python
# use month check column to check month column make sure no unmmatch error


(sales_df['month_check'] == sales_df['Month_name']).all()
```

```python
# drop excess columns

sales_df = sales_df.drop(columns=[
    'hour_of_day',
    'Weekday',
    'Month_name'
])
    
```

```python
sales_df
```

```python
# revenue by month 

monthly_revenues= sales_df.groupby('month_check')['money'].sum().sort_values(ascending=False)
```

```python
monthly_revenues
```

```python
# revenue by weekday

weekday_revenues = sales_df.groupby('weekday_check')['money'].sum().sort_values(ascending=False)
```

```python
weekday_revenues
```

```python
# revenue by coffee type

coffee_revenue = sales_df.groupby('coffee_name')['money'].sum().sort_values(ascending=False)
```

```python
coffee_revenue
```

```python
# Amount sold by coffee type

coffee_sales_count = sales_df.groupby('coffee_name').size().sort_values(ascending=False)
```

```python
coffee_sales_count
```

```python
# avg amount spent 

avg_purchase = round(sales_df['money'].mean(), 2)
```

```python
avg_purchase
```

```python
# revenue by the hour

hourly_revenue = sales_df.groupby('hour_check')['money'].sum().sort_values(ascending=False)
```

```python
hourly_revenue
```

```python
# payment type 

payment_type = sales_df.groupby('cash_type')['money'].sum()
```

```python
payment_type
```

```python
# check if only one payment type is there

sales_df['cash_type'].nunique()
```

```python
coffee_revenue.plot(kind='bar', color='brown', figsize=(10,5))

plt.xticks(rotation=45)

plt.title('Revenue by Coffee Type')
plt.xlabel('Coffee')
plt.ylabel('Revenue ($)')

plt.tight_layout()
plt.savefig('chart1_top_products.png')
plt.show()
```

```python
coffee_sales_count.plot(kind='bar', color='brown', figsize=(10,5))

plt.xticks(rotation=45)

plt.title('Amount of Coffee Sold')
plt.xlabel('Coffee')
plt.ylabel('Amount')

plt.tight_layout()
plt.savefig('chart2_top_products.png')
plt.show()
```

```python
month_order = ['March', 'April', 'May', 'June', 'July', 
               'August', 'September', 'October', 'November','December', 'January', 'February']

monthly_revenues.index = pd.Categorical(monthly_revenues.index, categories=month_order, ordered=True)

monthly_revenues = monthly_revenues.sort_index()


monthly_revenues.plot(kind='line', marker='o', color='brown', figsize=(10,5))

plt.xticks(range(len(month_order)), month_order, rotation=45) 
plt.title('Revenue by Month')
plt.xlabel('Months')
plt.ylabel('Revenue ($)')
plt.tight_layout()
plt.savefig('chart1_best_months.png')
plt.show()
```

```python
day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
weekday_revenues.reindex(day_order).plot(kind='line',marker = 'o', color='brown', figsize=(10,5))

plt.xticks(rotation=45)

plt.title('Revenue by Day of Week')
plt.xlabel('Day')
plt.ylabel('Revenue ($)')

plt.tight_layout()
plt.savefig('chart3_best_days.png')
plt.show()
```

Tuesday, Monday, and Friday are the most profitable days in that order. Sunday and saturday are the least profitable in that order, with weekends being significantly less profitable compared to the other days

Most customers for coffee shop are workers who work a monday to friday schedule.

To increase revenue on weekend it is recommended to draw more customers in through use of space. recommend ways to use space are hosting events such as small gathering, events for the coffee shop, drawing in customers and new regulars on weekend.

Elevating the space for weekend crowd is beneficial, outlets for those who wish to bring laptops to work in coffee shop. Promoting longer stay in the coffee shop.
Seating arrangement for promoting community between customers.

```python
hourly_revenue.sort_index().plot(kind='line', marker='o', color='brown', figsize=(10,5))

plt.title('Revenue by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Revenue ($)')

plt.tight_layout()
plt.savefig('chart2_best_hours.png')
plt.show()
```

Below is code comparing results by year. The data only contains from march 2024 to march 2025, making the data not impactful for client at the moment. In the future when more data is provided the code below can be used.

```python
# revenue by year, data cotains march 2024 to march 2025

revenue_by_year = sales_df.groupby(sales_df.datetime.dt.year)['money'].sum()
```

```python
revenue_by_year
```

```python
# coffee type revenue per year, data cotains march 2024 to march 2025

coffee_revenue_by_year = sales_df.groupby(['coffee_name', sales_df.datetime.dt.year])['money'].sum().unstack()
```

```python
coffee_revenue_by_year
```

```python
colors = ['#8B4513', '#D2B48C']

coffee_revenue_by_year.plot(kind='bar', color=colors, figsize=(10,5))

plt.xticks(rotation=45)

plt.title('Coffee Revenue Year Comparison')
plt.xlabel('Coffee')
plt.ylabel('Revenue ($)')

plt.tight_layout()
plt.savefig('chart3_top_products_comparison.png')
plt.show()
```

```python
# amount of coffee type sold per year, data cotains march 2024 to march 2025

coffee_sold_by_year = sales_df.groupby(['coffee_name', sales_df.datetime.dt.year]).size().unstack()
```

```python
coffee_sold_by_year
```

```python
colors = ['#8B4513', '#D2B48C']

coffee_sold_by_year.plot(kind='bar', color=colors, figsize=(10,5))

plt.xticks(rotation=45)

plt.title('Amount of Coffee Sold')
plt.xlabel('Coffee')
plt.ylabel('Amount')

plt.tight_layout()
plt.savefig('chart2_top_products_comparisons.png')
plt.show()
```
